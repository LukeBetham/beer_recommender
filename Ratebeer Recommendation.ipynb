```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_clean.csv")
```

```python
# adding numerical ID to reviewers and beers
df['beer_id'] = pd.factorize(df.beer_name)[0]
df['reviewer_id'] = pd.factorize(df.reviewer_name)[0]
```

```python
df.head(2)
```

## Using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_distances, cosine_similarity, pairwise_distances, euclidean_distances
from scipy import sparse
```

```python
row = df.reviewer_id
col = df.beer_id
data = df.beer_rating
rui = sparse.csr_matrix((data, (row, col)), shape=(df.reviewer_id.unique().shape[0], df.beer_id.unique().shape[0]))
rui
```

```python
user_similarity_cos = cosine_similarity(rui)
```

```python
df_sim = pd.DataFrame(user_similarity_cos)
```
